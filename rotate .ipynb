Import original image

```python
img = Image.open("/content/source.jpg") 
img
```

Method for rotating image, parameter 'num' indicates how many time of rotation.

```python
def rotate(img, num):
  for x in range(0, 360, 360/num):
    tmp = img.rotate(x)
    tmp.save(f'example_{x}.jpg')
    files.download(f'example_{x}.jpg') 
```

```python
rotate(img)
```
